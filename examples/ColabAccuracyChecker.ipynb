```python
%cd '/content/drive/My Drive/yoloneat-viz/'
import sys
import os
import glob
sys.path.append("../")
from yoloneatviz import HolovizNapari
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot  as plt
from natsort import natsorted
import numpy as np
```

```python
thresholdspace = 10
thresholdtime = 4
MovieName = []
NetworkName = []
CellName = []
Accuracy = []
resultsdir = '/content/drive/My Drive/Datasets/CAPED/Cherry/Results/'
Path(resultsdir).mkdir(exist_ok=True)

csv_gt = '/content/drive/My Drive/Datasets/CAPED/Cherry/CAPEDDivisionH2BmCherry_timelapse_60h.csv'

csv_pred = '/content/drive/My Drive/Datasets/CAPED/Cherry/Detection_d11_th1_iou0.2/DivisionLocationH2BmCherry_timelapse_60h.csv'
Name = os.path.basename(os.path.splitext(csv_pred)[0])
tp = HolovizNapari.TruePositives(csv_gt, csv_pred, thresholdspace = thresholdspace, thresholdtime = thresholdtime)
MovieName.append(Name)
NetworkName.append('Detection_d11_iou0.2_fi1')
CellName.append('Division')
Accuracy.append(str('%.4f'%tp))


csv_pred = '/content/drive/My Drive/Datasets/CAPED/Cherry/Detection_d29_th1_iou0.2/DivisionLocationH2BmCherry_timelapse_60h.csv'
Name = os.path.basename(os.path.splitext(csv_pred)[0])
tp = HolovizNapari.TruePositives(csv_gt, csv_pred, thresholdspace = thresholdspace, thresholdtime = thresholdtime)
MovieName.append(Name)
NetworkName.append('Detection_d29_iou0.2_fi1')
CellName.append('Division')
Accuracy.append(str('%.4f'%tp))

csv_pred = '/content/drive/My Drive/Datasets/CAPED/Cherry/Detection_d38_th1_iou0.2/DivisionLocationH2BmCherry_timelapse_60h.csv'
Name = os.path.basename(os.path.splitext(csv_pred)[0])
tp = HolovizNapari.TruePositives(csv_gt, csv_pred, thresholdspace = thresholdspace, thresholdtime = thresholdtime)
MovieName.append(Name)
NetworkName.append('Detection_d38_iou0.2_fi1')
CellName.append('Division')
Accuracy.append(str('%.4f'%tp))

df = pd.DataFrame(list(zip(MovieName,NetworkName,CellName,Accuracy)), index = None, 
                                                                      columns =['MovieName', 'NetworkName', 'CellType',  'Accuracy'])
                        
df.to_csv(resultsdir + '/' + 'DetectionAccuracy_iou0.2' +  '.csv')  
df

Accuracy = natsorted(Accuracy, reverse = True)

colors = np.random.rand(len(Accuracy))
plt.scatter(NetworkName, Accuracy,  alpha=0.5, c = colors)


plt.savefig(resultsdir + '/' + 'DetectionAccuracy_iou0.2' + '.png', dpi = 300)
plt.show()
df
```
